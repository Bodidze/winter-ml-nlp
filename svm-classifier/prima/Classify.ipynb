```python
import codecs
from mlclassifier import Categorizer

model_path = './models_categories_nltk'
classifier = Categorizer()
classifier.load_models(model_path)
print(classifier._tokenization)
```

```python
v = classifier.classify('Life of the animals is concentrated in areas near the coast.', 'SVC')
print(v)
v = classifier.classify('Life of the animals is concentrated in areas near the coast.', 'MultinomialNB')
print(v)
```

```python
v = classifier.classify('Are you ready to put those tequila-shot nightmares to bed?', 'SVC')
print(v)
v = classifier.classify('Are you ready to put those tequila-shot nightmares to bed?', 'MultinomialNB')
print(v)
```

```python
v = classifier.classify('A schools communication through website or mobile application.', 'SVC')
print(v)
v = classifier.classify('A schools communication through website or mobile application.', 'MultinomialNB')
print(v)
```

```python
v = classifier.classify('Marly used to like me, but he was a dirty nigga back then.', 'SVC')
print(v)
v = classifier.classify('Marly used to like me, but he was a dirty nigga back then.', 'MultinomialNB')
print(v)
```

```python
with codecs.open('./random_data.txt', encoding='utf-8', mode='r') as f:
    for line in f:
        vect = classifier.classify(line, 'SVC')
        print(vect)
```
