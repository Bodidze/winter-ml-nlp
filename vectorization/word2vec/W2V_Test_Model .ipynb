```python
from gensim.models import Word2Vec
# Load Word2Vec model from file.
model = Word2Vec.load('./w2v/w2v_model.bin')
```

```python
# Words similarity
print(model.similarity('woman', 'man'))
```

```python
# Most similar words.
print(model.most_similar(positive=['war', 'star'], negative=['sun']))
```
